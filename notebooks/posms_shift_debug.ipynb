```python
import pprint

# =========================
#  パス設定
# =========================
notebook_dir = Path().resolve()
project_root = notebook_dir.parent
csv_dir = project_root / "excel_templates" / "export_csv"

print("csv_dir:", csv_dir, "exists:", csv_dir.exists())

# =========================
#  シフト期間 D（28日分）を作る
#  shift_meta.csv の start_date を起点にする
# =========================
df_meta = pd.read_csv(csv_dir / "shift_meta.csv", encoding="cp932")
start_date = pd.to_datetime(df_meta.iloc[0]["start_date"]).normalize()

# 日曜日から始まる 28 日間（start_date が既に日曜日になっている想定）
D = pd.date_range(start=start_date, periods=28)
D_norm = D.normalize()
shift_set = set(D_norm)   # 範囲チェック用

# --- 祝日・平日マスク（必要なら使う） ---
is_holiday_mask = D_norm.map(jpholiday.is_holiday)
weekday_series = D_norm.weekday  # 0=Mon … 6=Sun

D_holiday  = D_norm[is_holiday_mask]
mask_weekday = (~is_holiday_mask) & (weekday_series < 5)
D_weekday = D_norm[mask_weekday]
mask_saturday = (~is_holiday_mask) & (weekday_series == 5)
D_saturday = D_norm[mask_saturday]
mask_sunday = (~is_holiday_mask) & (weekday_series == 6)
D_sunday = D_norm[mask_sunday]

print("D (28日):", D_norm, "\n")

# =========================
#  pre_assignments.csv（上段＋下段）
# =========================
df_pre = pd.read_csv(csv_dir / "pre_assignments.csv", encoding="cp932")

# 列を念のため揃える
df_pre = df_pre[["氏名", "日付", "行種別", "セル値"]]

# 日付を Timestamp 正規化
df_pre["日付"] = pd.to_datetime(df_pre["日付"]).dt.normalize()

# ★ シフト期間外の行は全部捨てる
df_pre = df_pre[df_pre["日付"].isin(shift_set)].copy()

print("df_pre rows (after filtering by D):", len(df_pre))

# =========================
#  special_marks.csv（廃休／マル超、色ベース）
# =========================
special_path = csv_dir / "special_marks.csv"

if special_path.exists():
    df_special = pd.read_csv(special_path, encoding="cp932")
    # 列揃え
    df_special = df_special[["氏名", "日付", "区分"]]
    df_special["日付"] = pd.to_datetime(df_special["日付"]).dt.normalize()
    # ★ シフト期間外の行は全部捨てる
    df_special = df_special[df_special["日付"].isin(shift_set)].copy()
else:
    df_special = pd.DataFrame({"氏名": [], "日付": [], "区分": []})

print("df_special rows (after filtering by D):", len(df_special))

# 休暇候補（leave_types.csv より）
leave_codes = set(
    df_leave["leave_name"]
    .astype(str)
    .str.strip()
    .tolist()
)

# 区名候補（zones.csv の "区名" 列を想定）
if "区名" in df_zone.columns:
    zone_codes = set(
        df_zone["区名"]
        .astype(str)
        .str.strip()
        .tolist()
    )
else:
    # 万一 "区名" 以外の列名なら、print(df_zone.columns) で確認して差し替え
    print("df_zone.columns =", df_zone.columns)
    zone_codes = set()

# =========================
#  コードリスト
# =========================
# 下段の休暇コードとして扱うもの
pre_list_rest    = ["R"]           # 必要に応じて調整
# 下段の勤務コードとして扱うもの
pre_list_work    = ["K"]           # 必要に応じて調整
# 特別業務（廃休/マル超）
pre_list_special = ["廃休", "マル超"]

# =========================
#  辞書の初期化
# =========================
pre_dict_rest    = {}  # {(氏名, 日付): 休暇コード}
pre_dict_work    = {}  # {(氏名, 日付): 勤務コード}
pre_dict_special = {}  # {(氏名, 日付): 特別業務コード}
pre_dict_time    = {}  # {(氏名, 日付): 時間帯コード（日1など）}

# =========================
#  df_pre（上段＋下段）から辞書構築
# =========================
for _, row in df_pre.iterrows():
    name     = row["氏名"]
    date_key = row["日付"]
    kind     = row["行種別"]    # "上段" or "下段"
    val      = row["セル値"]

    if pd.isna(val) or str(val).strip() == "":
        continue

    val_str = str(val).strip()

    # --- 上段：時間帯コード（日1 など） ---
    if kind == "上段":
        pre_dict_time[(name, date_key)] = val_str

    # --- 下段：勤務(区) / 休暇 / 廃休 / マル超 / その他 ---
    elif kind == "下段":
        # 特別業務（廃休／マル超）
        if val_str in pre_list_special:
            pre_dict_special[(name, date_key)] = val_str

        # 休暇（非番／週休／年休／…）
        elif val_str in leave_codes:    # または pre_list_rest
            pre_dict_rest[(name, date_key)] = val_str

        # 勤務（区名：1区／2区／…）
        elif val_str in zone_codes:     # または pre_list_work
            pre_dict_work[(name, date_key)] = val_str

        else:
            # それ以外（備考・特殊文字など）
            # 必要なら別の dict に振り分ける：
            # pre_dict_other[(name, date_key)] = val_str
            pass

# =========================
#  df_special（色ベースの廃休／マル超）で上書き
#   → 色指定を優先したい場合
# =========================
for _, row in df_special.iterrows():
    name     = row["氏名"]
    date_key = row["日付"]
    kind     = str(row["区分"]).strip()   # "廃休" or "マル超"

    if kind in pre_list_special:
        pre_dict_special[(name, date_key)] = kind

# =========================
#  デバッグ出力
# =========================
print("■ 事前指定休暇 (pre_dict_rest)")
pprint.pprint(pre_dict_rest)
print("\n■ 事前指定勤務 (pre_dict_work)")
pprint.pprint(pre_dict_work)
print("\n■ 事前指定特別業務 (pre_dict_special)")
pprint.pprint(pre_dict_special)
print("\n■ 事前指定時間帯 (pre_dict_time)")
pprint.pprint(pre_dict_time)
```

```python
# 1) Python の weekday() から曜日ラベルへのマッピング
weekday_map = {0: '月', 1: '火', 2: '水', 3: '木', 4: '金', 5: '土', 6: '日'}

# 2) 各日付 d∈D を「月～日 or 祝」に変換
date_label = {}
for d in D:
    if jpholiday.is_holiday(d):      # 祝日なら '祝'
        date_label[d] = '祝'
    else:
        date_label[d] = weekday_map[d.weekday()]
#各社員が出勤できる曜日を1、出勤しない(ソルバーが選ばない曜日)を0に設定
# 3) avail[(i,d)] = 1 or 0 の辞書を作成　
#事前指定で出勤しない日に指定が入っていたら1にする
avail = {
    (i, d): 1 if (i, d) in pre_dict_work else C[i][ date_label[d] ]
    for i in C
    for d in D
}
```

```python
req = {}  # {(pd.Timestamp, 業務種類): 1}

for d in D:
    # d は pd.Timestamp
    # 祝日判定には d.date()（datetime.date）を渡す
    if jpholiday.is_holiday(d.date()):
        label = '祝'
    else:
        wd = d.weekday()  # 0=月…6=日
        if wd < 5:
            label = ['月','火','水','木','金'][wd]
        else:
            label = '土' if wd == 5 else '日'

    # mapping[label] の各業務種類をすべて需要１として登録
    for k in mapping.get(label, []):
        req[(d, k)] = 1
```

```python
import pulp
```

```python
#数理モデルのインスタンス作成
model = pulp.LpProblem("Shift_Scheduling", pulp.LpMinimize)
```

```python
# バイナリ変数の定義
x = pulp.LpVariable.dicts(
    "Shift",
    [
        (i, d, k)
        for i in I
        for d in D
        for k in K
    ],
    cat='Binary'
)
```

```python
# 補助変数 y[i,d] = 1 なら社員iが日にちdに勤務
y = pulp.LpVariable.dicts(
    "WorkDay",
    [(i, d) for i in I for d in D],
    cat='Binary'
)
```

```python
# 休みタイプ変数 rest[i,d,r]　必要
rest = pulp.LpVariable.dicts(
    "Rest",
    [(i, d, r) for i in I for d in D for r in R],
    cat='Binary'
)
```

```python
specialWork = pulp.LpVariable.dicts(
    "X_Special",
    [
        (i,d,s)
        for i in I
        for d in D
        for s in special_attendance  # (廃休 or マル超)
    ],
    cat=pulp.LpBinary
)
```

```python
#欠区の定義
# missing=1 なら (d,t,k) は欠区（割り当て0人）
# missing=0 なら 需要をフルに満たす
missing = pulp.LpVariable.dicts(
    "Missing",
    [(d,k) for (d,k) in req.keys()],
    cat='Binary'
)
```

```python
#連勤を判定するための補助変数
normal_work = pulp.LpVariable.dicts(
    "NormalWork",
    [(i,d) for i in I for d in D],
    cat=pulp.LpBinary
)
```

```python
#曜日による出勤の指定
#事前指定より先に実行して、事前指定があれば上書きされるようにする
for i in I:
    for d in D:
        for k in K:
            # avail[(i,d)] が 0(出勤しない曜日) のとき x[(i,d,k)] == 0
            model += x[(i, d, k)] <= avail[(i, d)], f"Avail_{i}_{d}_{k}"
```

```python
# ---------- 休暇の事前指定 ----------
for i in I:
    for d in D:
        # ----- 休暇指定がある場合 (事前指定で休暇にする日) -----
        if (i, d) in pre_dict_rest:
            rest_type = pre_dict_rest[(i, d)]
            model += rest[(i, d, rest_type)] == 1, f"Rest_{i}_{d}_{rest_type}"
           # 他の通常休暇は0
            for r_ in R:
                if r_ != rest_type:
                    model += rest[(i, d, r_)] == 0, f"RestZero_{i}_{d}_{r_}" 

            # 休暇指定の日は勤務しない → x=0, specialWork=0
            for k in K:
                model += x[(i, d, k)] == 0, f"ZeroX_{i}_{d}_{k}"
                
            # 特別状態(廃休,マル超) も0
            for s in special_attendance:
                model += specialWork[(i, d, s)] == 0, f"ZeroSpWork_{i}_{d}_{s}"

        else:
            # ----- 休暇指定がない → 勤務の事前指定をチェック ----
            # 休暇指定がない
            # ソルバーに選ばせたくない休暇 (例: "計年","夏休"...) は 0固定
            # 逆に "週休","非番","祝休"などはソルバーが使うかもしれない(要件次第)
            for r in R:
                if r not in ["週休", "非番", "祝休"]:
                    model += rest[(i, d, r)] == 0, f"ZeroRest_{i}_{d}_{r}"
```

```python
#"""廃休、マル超の事前指定"""
for i in I:
    for d in D:
        if (i, d) in pre_dict_special:
            s_type = pre_dict_special[(i, d)]
            for s in special_attendance:
                if s == s_type:
                    model += specialWork[(i, d, s)] == 1, f"FixSpec_{i}_{d}_{s}"
                else:
                    model += specialWork[(i, d, s)] == 0, f"ZeroSpec_{i}_{d}_{s}"

        else:
            # 特別状態の指定なし -> 全0
            for s in special_attendance:
                model += specialWork[(i, d, s)] == 0, f"NoSpec_{i}_{d}_{s}"
```

```python
#"""業務の事前指定"""
for i in I:
    for d in D:
        for k in K:
            # もし (i, d) が pre_dict_work にあって、
            # その値がこの k なら x[(i,d,k)] == 1
            if pre_dict_work.get((i, d)) == k:
                model += x[(i, d, k)] == 1, f"ForceWork_{i}_{d}_{k}"
            else:
                # req に無いなら 0 固定
                if (d, k) not in req:
                    model += x[(i, d, k)] == 0, f"NoReq_{i}_{d}_{k}"
                else:
                    # それ以外はソルバー自由
                    pass
```

```python
# 各社員が各日にちに勤務(y[i,d])するか、休暇(rest[i,d,r])を取るか、
for i in I:
    for d in D:
        model += (
            y[i, d]
            + pulp.lpSum(rest[i, d, r] for r in R)
        ) == 1, f"WorkOrRest_{i}_{d}"
```

```python
#"""１日に入れるのは１シフトのみ"""
for i in I:
    for d in D:
        # シフト×業務 の合計
        sum_x_id = pulp.lpSum(x[(i, d, k)] for k in K)

        # 1) 勤務するなら sum_x_id = 1
        # 2) 勤務しない(=0)なら sum_x_id = 0
        #  → sum_x_id == y[i,d]　1日出勤フラグの紐付け(重要)
        model += sum_x_id == y[(i,d)], f"OneShiftOrNone_{i}_{d}"
```

```python
for i in I:
    for d in D:
        for k in K:
            # 1) A に含まれる業務なら、ソルバー自由
            #    ただし (i,d) に対する事前指定がこの k なら強制 x=1
            if k in A[i]:
                if pre_dict_work.get((i, d)) == k:
                    model += x[(i, d, k)] == 1, f"MustWork_{i}_{d}_{k}"
                else:
                    # ソルバー自由 => 制約は書かない
                    pass

            # 2) B に含まれる業務なら、事前指定がこの k なら x=1、されていなければ x=0
            elif k in B[i]:
                if pre_dict_work.get((i, d)) == k:
                    model += x[(i, d, k)] == 1, f"MustWork_{i}_{d}_{k}"
                else:
                    model += x[(i, d, k)] == 0, f"NoWork_{i}_{d}_{k}"

            # 3) どちらにもない業務は担当不可
            else:
                model += x[(i, d, k)] == 0, f"InvalidJob_{i}_{d}_{k}"
```

```python
for i in I:
    for d in D:
        # 週休 + 廃休 <= 1
        model += rest[(i,d,'週休')] + specialWork[(i,d,'廃休')] <= 1
        # 非番 + マル超 <= 1
        model += rest[(i,d,'非番')] + specialWork[(i,d,'マル超')] <= 1
```

```python
for i in I:
    # 週休は月に4日
    model += (
        pulp.lpSum(
            rest[(i,d,'週休')] + specialWork[(i,d,'廃休')]
            for d in D
        ) == 4
    ), f"WeeklyOff_{i}"

    # 非番は4日以上
    model += (
        pulp.lpSum(
            rest[(i,d,'非番')] + specialWork[(i,d,'マル超')]
            for d in D
        ) >= 4
    ), f"Hiban_{i}"
```

```python
#週に1回週休をとる(廃休をとってもカウントされる)
for i in I:
    for w, days_in_week in enumerate(W):
        #rest(週休) + x_special(廃休) == 1
        model += (
            pulp.lpSum(rest[(i,d,'週休')] + specialWork[(i,d,'廃休')]
                        for d in days_in_week) == 1
        )
```

```python
#スラック変数
devPos = {}
devNeg = {}
```

```python
#なるべく週に１回非番をとるがとれない場合はスラック変数により別の週にとることも可
for i in I:
    for w, days_in_week in enumerate(W):
        # 週の "(非番 + マル超)" 回数
        sumHibanOrMarucho = pulp.lpSum(
            rest[(i,d,'非番')] + specialWork[(i,d,'マル超')]
            for d in days_in_week
        )
        
        # スラック変数
        devPos[(i,w)] = pulp.LpVariable(f"devPos_{i}_{w}", lowBound=0)
        devNeg[(i,w)] = pulp.LpVariable(f"devNeg_{i}_{w}", lowBound=0)
        
        model += sumHibanOrMarucho - 1 <= devPos[(i,w)]
        model += 1 - sumHibanOrMarucho <= devNeg[(i,w)]
```

```python
#日曜日に休む場合は週休
for i in I:
    for d in D_sunday:
        model += y[i,d] + rest[i,d,'週休'] == 1, f"SundayWorkOrWeeklyOff_{i}_{d}"
```

```python
#土曜日に休む場合は非番
for i in I:
    for d in D_saturday:
        # もし (i,d) に特別休暇が指定されていればスキップ
        if (i, d) in pre_dict_rest:
            continue
        
        # 指定の休みが無い => 土曜は (出勤 or 非番)
        model += (
            y[(i,d)] + rest[(i,d,'非番')]
            == 1
        ), f"SaturdayWorkOrNonban_{i}_{d}"
```

```python
#祝日以外の日は祝休を取れない
for i in I:
    for d in D:
        if d not in D_holiday:
            # 祝日以外の日 => 祝日休 は 0
            model += rest[i,d,'祝休'] == 0, f"NoHolidayRest_{i}_{d}"
```

```python
# 祝日に非番の指定
for d in D_holiday:
    for i in I:
        #非番の指定があれば入れられる。
        if (i,d) in pre_dict_rest and pre_dict_rest[(i,d)] == '非番':
            continue
        #指定がなければ祝日は "出勤 or 祝休" のどちらか
        else:
            model += y[(i,d)] + rest[(i,d,'祝休')] == 1
```

```python
#"""連勤に関する制約"""
#10連勤の禁止　廃休、マル超を取れば９連勤まで可
for i in I:
    # d_idxを 0..(28-10) までループ => 0..18
    for d_idx in range(len(D) - 10 + 1):  # 28-10+1=19
        # d_idx..d_idx+9 (10日分)
        consecutive_days = [D[d_idx + offset] for offset in range(10)]
        
        # y[(i, day)] を連続10日分集計
        model += (
            pulp.lpSum(y[(i, day_)] for day_ in consecutive_days)
            <= 9
        ), f"No10ConsecutiveDays_{i}_{d_idx}"
                
#紐付け制約: normal_work <= y, normal_work <= 1 - x_special
for i in I:
    for d in D:
        model += normal_work[(i,d)] >= y[(i,d)] - specialWork[(i,d,'廃休')] - specialWork[(i,d,'マル超')]
        model += normal_work[(i,d)] <= 1 - specialWork[(i,d,'廃休')]
        model += normal_work[(i,d)] <= 1 - specialWork[(i,d,'マル超')]
        

#廃休、マル超を取らない連勤は５まで。
for i in I:
    for start_idx in range(len(D) - 6 + 1):  # 0..22
        consecutive_dates = [D[start_idx + r] for r in range(6)]
        # ここで normal_work[(i, consecutive_dates[r])] を合計
        model += (
            pulp.lpSum(normal_work[(i, d_)] for d_ in consecutive_dates)
            <= 5
        )
```

```python
#"""社員の出勤可能数と業務数の過不足を計算"""
# 1) 社員ごとの出勤可能日数を計算
# 1) 社員ごとの出勤可能日数を、C の可否から直接カウント
base = 20  # 最大勤務日数の上限

# avail[(i, d)] が 1 の日を数えて possible を計算し、上限をかける
employee_possible_days = {
    i: min(
        sum(avail[(i, d)] for d in D),
        base
    )
    for i in I
}

# 確認出力
for i, cnt in employee_possible_days.items():
    print(f"{i} の出勤可能日数: {cnt} 日")
for i in I:
    extra_rest_count = 0
        
    # (1-a) 特別休暇を差し引く
    for d in D:
        if (i, d) in pre_dict_rest:
            r_type = pre_dict_rest[(i, d)]
            if r_type in ['計年', '年休', '夏期', '冬期', '代休',
                            '承欠', '休職', '産休', '育休', '介護', '病休', 'その他']:
                extra_rest_count += 1
    # 仮での基本出勤可能日数
    employee_possible_days[i] -= extra_rest_count

    # (1-b) 廃休 or マル超 指定があれば +1
    for d in D:
        if (i, d) in pre_dict_special:
            s_type = pre_dict_special[(i,d)]
            if s_type in ['廃休','マル超']:
                employee_possible_days[i] += 1

# (1-c) 出張や研修など (reqにない業務) の事前指定があれば -1
for (i, d), k in pre_dict_work.items():
    if (d, k) not in req:
        # この業務に入ってもreqは満たせないため-1
        employee_possible_days[i] -= 1

# 2) 需要計算
# 2) 需要計算：req による実際の要出勤数を集計
#バランス調整用の補助20を引く
total_demand = sum(req.values()) - 20

print(f"【需要計算】total_demand = {total_demand} シフト枠")

#3) 供給計算
# 例: 全社員の出勤可能日数を合計
total_supply = sum(employee_possible_days.values())

# 4) 過不足を計算
excess_or_deficiency = total_supply - total_demand
print(f"【出勤可能日数計算】")
for i in I:
    print(f"  - {i}: {employee_possible_days[i]}日")

print(f"total_supply={total_supply}")
# あるいはゼロを +0 と表示したくない場合は条件分岐で制御
sign = '+' if excess_or_deficiency > 0 else ''
print(f"過不足 = {sign}{excess_or_deficiency}")
```
