```python
from pathlib import Path
import pandas as pd
import jpholiday
import pprint

pp = pprint.PrettyPrinter(width=120)

# =========================
#  CSV読み込み（文字コード自動）
# =========================
def read_csv_auto(path: Path, **kwargs) -> pd.DataFrame:
    last_err = None
    for enc in ("utf-8-sig", "cp932"):
        try:
            return pd.read_csv(path, encoding=enc, **kwargs)
        except UnicodeDecodeError as e:
            last_err = e
    raise last_err

def must_have_cols(df: pd.DataFrame, need: set[str], name: str):
    missing = need - set(df.columns)
    if missing:
        raise ValueError(f"{name}: missing columns {missing}. got={df.columns.tolist()}")

# =========================
#  パス設定
# =========================
notebook_dir = Path().resolve()
project_root = notebook_dir.parent
csv_dir = project_root / "excel_templates" / "export_csv"
print("csv_dir:", csv_dir, "exists:", csv_dir.exists())
if not csv_dir.exists():
    raise FileNotFoundError(csv_dir)

# =========================
#  shift_meta.csv → start_date → 28日 D を作る
# =========================
df_meta = read_csv_auto(csv_dir / "shift_meta.csv")
must_have_cols(df_meta, {"start_date"}, "shift_meta.csv")

start_date = pd.to_datetime(df_meta.loc[0, "start_date"]).normalize()
D = pd.date_range(start=start_date, periods=28).normalize()
shift_set = set(D)

print("start_date:", start_date)
print("D (28日):", D)

# --- 祝日・平日マスク（必要なら） ---
is_holiday_mask = pd.Series([jpholiday.is_holiday(d) for d in D], index=D)
weekday_series = pd.Series([d.weekday() for d in D], index=D)  # 0=Mon … 6=Sun

D_holiday   = D[is_holiday_mask.values]               # 祝日（※日曜も True になり得るが、jpholiday次第）
D_weekday   = D[(~is_holiday_mask).values & (weekday_series.values < 5)]
D_saturday  = D[(~is_holiday_mask).values & (weekday_series.values == 5)]
D_sunday    = D[(~is_holiday_mask).values & (weekday_series.values == 6)]

# =========================
#  employees.csv（emp_no→氏名の辞書：デバッグ用）
# =========================
df_emp = read_csv_auto(csv_dir / "employees.csv")

# emp_no列と氏名列は環境で列名が揺れることがあるので候補で探す
emp_no_col_candidates = ["emp_no", "社員番号", "社員No", "社員NO", "従業員番号"]
emp_name_col_candidates = ["name", "氏名", "社員名", "名前"]

def find_col(df, candidates):
    for c in candidates:
        if c in df.columns:
            return c
    return None

emp_no_col = find_col(df_emp, emp_no_col_candidates)
emp_name_col = find_col(df_emp, emp_name_col_candidates)

empno_to_name = {}
if emp_no_col is not None and emp_name_col is not None:
    s_emp = df_emp[[emp_no_col, emp_name_col]].copy()
    s_emp[emp_no_col] = s_emp[emp_no_col].astype(str).str.strip()
    s_emp[emp_name_col] = s_emp[emp_name_col].astype(str).str.strip()
    empno_to_name = dict(zip(s_emp[emp_no_col], s_emp[emp_name_col]))
else:
    print("NOTE: employees.csv から emp_no/氏名列が見つからず、名前表示は省略します。")
    print("employees.csv columns =", df_emp.columns.tolist())

# =========================
#  leave_types.csv → leave_codes
# =========================
df_leave = read_csv_auto(csv_dir / "leave_types.csv")
# 期待列は leave_name（VBA/EXE想定）
if "leave_name" not in df_leave.columns:
    # 代替候補
    alt = find_col(df_leave, ["休暇種類名", "休暇名"])
    if alt is None:
        raise ValueError(f"leave_types.csv: leave_name列が見つかりません。columns={df_leave.columns.tolist()}")
    df_leave = df_leave.rename(columns={alt: "leave_name"})

leave_codes = set(df_leave["leave_name"].astype(str).str.strip().tolist())

# =========================
#  zones.csv → zone_codes
# =========================
df_zone = read_csv_auto(csv_dir / "zones.csv")
zone_name_col = find_col(df_zone, ["区名", "zone_name", "name"])
if zone_name_col is None:
    print("zones.csv columns =", df_zone.columns.tolist())
    raise ValueError("zones.csv に区名列が見つかりません（候補: 区名/zone_name/name）")

zone_codes = set(df_zone[zone_name_col].astype(str).str.strip().tolist())

# =========================
#  pre_assignments.csv（上段＋下段）
#  形式: emp_no,date,row_kind,value
# =========================
df_pre = read_csv_auto(csv_dir / "pre_assignments.csv")
must_have_cols(df_pre, {"emp_no", "date", "row_kind", "value"}, "pre_assignments.csv")

df_pre["emp_no"] = df_pre["emp_no"].astype(str).str.strip()
df_pre["date"] = pd.to_datetime(df_pre["date"]).dt.normalize()
df_pre["row_kind"] = df_pre["row_kind"].astype(str).str.strip()
df_pre["value"] = df_pre["value"].astype(str).str.strip()

# 期間外は捨てる
df_pre = df_pre[df_pre["date"].isin(shift_set)].copy()
print("df_pre rows (after filtering by D):", len(df_pre))

# =========================
#  special_marks.csv（廃休/マル超）
#  形式: emp_no,date,kind
# =========================
special_path = csv_dir / "special_marks.csv"
if special_path.exists():
    df_special = read_csv_auto(special_path)
    must_have_cols(df_special, {"emp_no", "date", "kind"}, "special_marks.csv")

    df_special["emp_no"] = df_special["emp_no"].astype(str).str.strip()
    df_special["date"] = pd.to_datetime(df_special["date"]).dt.normalize()
    df_special["kind"] = df_special["kind"].astype(str).str.strip()

    df_special = df_special[df_special["date"].isin(shift_set)].copy()
else:
    df_special = pd.DataFrame({"emp_no": [], "date": [], "kind": []})

print("df_special rows (after filtering by D):", len(df_special))

# =========================
#  コードリスト（ここは“現行の仕様”に寄せる）
# =========================
SPECIAL_KINDS = {"廃休", "マル超"}  # 特別業務

# =========================
#  辞書の初期化（すべて emp_no キー）
# =========================
pre_dict_rest    = {}  # {(emp_no, date): 休暇コード}
pre_dict_work    = {}  # {(emp_no, date): 勤務コード(区名など)}
pre_dict_special = {}  # {(emp_no, date): 廃休/マル超}
pre_dict_time    = {}  # {(emp_no, date): 上段（時間帯等）}

# “未知コード”を拾っておく（入力ミス検出に使う）
pre_unknown_lower = []  # [(emp_no, date, value)]

# =========================
#  pre_assignments から辞書構築
# =========================
for _, row in df_pre.iterrows():
    emp = row["emp_no"]
    d   = row["date"]
    kind = row["row_kind"]   # "upper" / "lower"
    v = row["value"].strip()

    if v == "" or v.lower() == "nan":
        continue

    if kind == "upper":
        pre_dict_time[(emp, d)] = v

    elif kind == "lower":
        if v in SPECIAL_KINDS:
            pre_dict_special[(emp, d)] = v
        elif v in leave_codes:
            pre_dict_rest[(emp, d)] = v
        elif v in zone_codes:
            pre_dict_work[(emp, d)] = v
        else:
            pre_unknown_lower.append((emp, d, v))

# =========================
#  special_marks（色ベース）を優先して上書き
# =========================
for _, row in df_special.iterrows():
    emp = row["emp_no"]
    d   = row["date"]
    k   = row["kind"]
    if k in SPECIAL_KINDS:
        pre_dict_special[(emp, d)] = k

# =========================
#  デバッグ表示（emp_no→氏名があれば併記）
# =========================
def fmt_key(k):
    emp, d = k
    nm = empno_to_name.get(emp, "")
    if nm:
        return f"{emp}({nm})", d.strftime("%Y-%m-%d")
    return emp, d.strftime("%Y-%m-%d")

def pretty_dict(dct, title):
    print(f"\n■ {title}  count={len(dct)}")
    # 見やすいようにキー整形して表示
    out = {(fmt_key(k)): v for k, v in dct.items()}
    pp.pprint(out)

pretty_dict(pre_dict_rest, "事前指定休暇 (pre_dict_rest)")
pretty_dict(pre_dict_work, "事前指定勤務 (pre_dict_work)")
pretty_dict(pre_dict_special, "事前指定特別業務 (pre_dict_special)")
pretty_dict(pre_dict_time, "事前指定時間帯 (pre_dict_time)")

if pre_unknown_lower:
    print("\n■ 下段の未知コード（入力ミス候補） count=", len(pre_unknown_lower))
    # 先頭だけ表示
    for emp, d, v in pre_unknown_lower[:30]:
        nm = empno_to_name.get(emp, "")
        emp_show = f"{emp}({nm})" if nm else emp
        print(emp_show, d.strftime("%Y-%m-%d"), v)
```

```python
def is_newyear_3days(d: pd.Timestamp) -> bool:
    # 例：1/1〜1/3
    return (d.month == 1 and d.day in (1, 2, 3))

def label_for_req(d: pd.Timestamp) -> str:
    wd = d.weekday()  # 0=月…6=日

    # 1) 正月3日間は祝日扱い（最優先）
    if is_newyear_3days(d):
        return "祝"

    # 2) 日曜は普通の日曜扱い（祝日より優先）
    if wd == 6:
        return "日"

    # 3) それ以外の祝日
    if jpholiday.is_holiday(d.date()):
        return "祝"

    # 4) 平日/土
    if wd < 5:
        return ["月", "火", "水", "木", "金"][wd]
    return "土"
```

```python
import pulp
```

```python
#数理モデルのインスタンス作成
model = pulp.LpProblem("Shift_Scheduling", pulp.LpMinimize)
```

```python
# バイナリ変数の定義
x = pulp.LpVariable.dicts(
    "Shift",
    [
        (i, d, k)
        for i in I
        for d in D
        for k in K
    ],
    cat='Binary'
)
```

```python
# 補助変数 y[i,d] = 1 なら社員iが日にちdに勤務
y = pulp.LpVariable.dicts(
    "WorkDay",
    [(i, d) for i in I for d in D],
    cat='Binary'
)
```

```python
# 休みタイプ変数 rest[i,d,r]　必要
rest = pulp.LpVariable.dicts(
    "Rest",
    [(i, d, r) for i in I for d in D for r in R],
    cat='Binary'
)
```

```python
specialWork = pulp.LpVariable.dicts(
    "X_Special",
    [
        (i,d,s)
        for i in I
        for d in D
        for s in special_attendance  # (廃休 or マル超)
    ],
    cat=pulp.LpBinary
)
```

```python
#欠区の定義
# missing=1 なら (d,t,k) は欠区（割り当て0人）
# missing=0 なら 需要をフルに満たす
missing = pulp.LpVariable.dicts(
    "Missing",
    [(d,k) for (d,k) in req.keys()],
    cat='Binary'
)
```

```python
#連勤を判定するための補助変数
normal_work = pulp.LpVariable.dicts(
    "NormalWork",
    [(i,d) for i in I for d in D],
    cat=pulp.LpBinary
)
```

```python
#曜日による出勤の指定
#事前指定より先に実行して、事前指定があれば上書きされるようにする
for i in I:
    for d in D:
        for k in K:
            # avail[(i,d)] が 0(出勤しない曜日) のとき x[(i,d,k)] == 0
            model += x[(i, d, k)] <= avail[(i, d)], f"Avail_{i}_{d}_{k}"
```

```python
# ---------- 休暇の事前指定 ----------
for i in I:
    for d in D:
        # ----- 休暇指定がある場合 (事前指定で休暇にする日) -----
        if (i, d) in pre_dict_rest:
            rest_type = pre_dict_rest[(i, d)]
            model += rest[(i, d, rest_type)] == 1, f"Rest_{i}_{d}_{rest_type}"
           # 他の通常休暇は0
            for r_ in R:
                if r_ != rest_type:
                    model += rest[(i, d, r_)] == 0, f"RestZero_{i}_{d}_{r_}" 

            # 休暇指定の日は勤務しない → x=0, specialWork=0
            for k in K:
                model += x[(i, d, k)] == 0, f"ZeroX_{i}_{d}_{k}"
                
            # 特別状態(廃休,マル超) も0
            for s in special_attendance:
                model += specialWork[(i, d, s)] == 0, f"ZeroSpWork_{i}_{d}_{s}"

        else:
            # ----- 休暇指定がない → 勤務の事前指定をチェック ----
            # 休暇指定がない
            # ソルバーに選ばせたくない休暇 (例: "計年","夏休"...) は 0固定
            # 逆に "週休","非番","祝休"などはソルバーが使うかもしれない(要件次第)
            for r in R:
                if r not in ["週休", "非番", "祝休"]:
                    model += rest[(i, d, r)] == 0, f"ZeroRest_{i}_{d}_{r}"
```

```python
#"""廃休、マル超の事前指定"""
for i in I:
    for d in D:
        if (i, d) in pre_dict_special:
            s_type = pre_dict_special[(i, d)]
            for s in special_attendance:
                if s == s_type:
                    model += specialWork[(i, d, s)] == 1, f"FixSpec_{i}_{d}_{s}"
                else:
                    model += specialWork[(i, d, s)] == 0, f"ZeroSpec_{i}_{d}_{s}"

        else:
            # 特別状態の指定なし -> 全0
            for s in special_attendance:
                model += specialWork[(i, d, s)] == 0, f"NoSpec_{i}_{d}_{s}"
```

```python
#"""業務の事前指定"""
for i in I:
    for d in D:
        for k in K:
            # もし (i, d) が pre_dict_work にあって、
            # その値がこの k なら x[(i,d,k)] == 1
            if pre_dict_work.get((i, d)) == k:
                model += x[(i, d, k)] == 1, f"ForceWork_{i}_{d}_{k}"
            else:
                # req に無いなら 0 固定
                if (d, k) not in req:
                    model += x[(i, d, k)] == 0, f"NoReq_{i}_{d}_{k}"
                else:
                    # それ以外はソルバー自由
                    pass
```

```python
# 各社員が各日にちに勤務(y[i,d])するか、休暇(rest[i,d,r])を取るか、
for i in I:
    for d in D:
        model += (
            y[i, d]
            + pulp.lpSum(rest[i, d, r] for r in R)
        ) == 1, f"WorkOrRest_{i}_{d}"
```

```python
#"""１日に入れるのは１シフトのみ"""
for i in I:
    for d in D:
        # シフト×業務 の合計
        sum_x_id = pulp.lpSum(x[(i, d, k)] for k in K)

        # 1) 勤務するなら sum_x_id = 1
        # 2) 勤務しない(=0)なら sum_x_id = 0
        #  → sum_x_id == y[i,d]　1日出勤フラグの紐付け(重要)
        model += sum_x_id == y[(i,d)], f"OneShiftOrNone_{i}_{d}"
```

```python
for i in I:
    for d in D:
        for k in K:
            # 1) A に含まれる業務なら、ソルバー自由
            #    ただし (i,d) に対する事前指定がこの k なら強制 x=1
            if k in A[i]:
                if pre_dict_work.get((i, d)) == k:
                    model += x[(i, d, k)] == 1, f"MustWork_{i}_{d}_{k}"
                else:
                    # ソルバー自由 => 制約は書かない
                    pass

            # 2) B に含まれる業務なら、事前指定がこの k なら x=1、されていなければ x=0
            elif k in B[i]:
                if pre_dict_work.get((i, d)) == k:
                    model += x[(i, d, k)] == 1, f"MustWork_{i}_{d}_{k}"
                else:
                    model += x[(i, d, k)] == 0, f"NoWork_{i}_{d}_{k}"

            # 3) どちらにもない業務は担当不可
            else:
                model += x[(i, d, k)] == 0, f"InvalidJob_{i}_{d}_{k}"
```

```python
for i in I:
    for d in D:
        # 週休 + 廃休 <= 1
        model += rest[(i,d,'週休')] + specialWork[(i,d,'廃休')] <= 1
        # 非番 + マル超 <= 1
        model += rest[(i,d,'非番')] + specialWork[(i,d,'マル超')] <= 1
```

```python
for i in I:
    # 週休は月に4日
    model += (
        pulp.lpSum(
            rest[(i,d,'週休')] + specialWork[(i,d,'廃休')]
            for d in D
        ) == 4
    ), f"WeeklyOff_{i}"

    # 非番は4日以上
    model += (
        pulp.lpSum(
            rest[(i,d,'非番')] + specialWork[(i,d,'マル超')]
            for d in D
        ) >= 4
    ), f"Hiban_{i}"
```

```python
#週に1回週休をとる(廃休をとってもカウントされる)
for i in I:
    for w, days_in_week in enumerate(W):
        #rest(週休) + x_special(廃休) == 1
        model += (
            pulp.lpSum(rest[(i,d,'週休')] + specialWork[(i,d,'廃休')]
                        for d in days_in_week) == 1
        )
```

```python
#スラック変数
devPos = {}
devNeg = {}
```

```python
#なるべく週に１回非番をとるがとれない場合はスラック変数により別の週にとることも可
for i in I:
    for w, days_in_week in enumerate(W):
        # 週の "(非番 + マル超)" 回数
        sumHibanOrMarucho = pulp.lpSum(
            rest[(i,d,'非番')] + specialWork[(i,d,'マル超')]
            for d in days_in_week
        )
        
        # スラック変数
        devPos[(i,w)] = pulp.LpVariable(f"devPos_{i}_{w}", lowBound=0)
        devNeg[(i,w)] = pulp.LpVariable(f"devNeg_{i}_{w}", lowBound=0)
        
        model += sumHibanOrMarucho - 1 <= devPos[(i,w)]
        model += 1 - sumHibanOrMarucho <= devNeg[(i,w)]
```

```python
#日曜日に休む場合は週休
for i in I:
    for d in D_sunday:
        model += y[i,d] + rest[i,d,'週休'] == 1, f"SundayWorkOrWeeklyOff_{i}_{d}"
```

```python
#土曜日に休む場合は非番
for i in I:
    for d in D_saturday:
        # もし (i,d) に特別休暇が指定されていればスキップ
        if (i, d) in pre_dict_rest:
            continue
        
        # 指定の休みが無い => 土曜は (出勤 or 非番)
        model += (
            y[(i,d)] + rest[(i,d,'非番')]
            == 1
        ), f"SaturdayWorkOrNonban_{i}_{d}"
```

```python
#祝日以外の日は祝休を取れない
for i in I:
    for d in D:
        if d not in D_holiday:
            # 祝日以外の日 => 祝日休 は 0
            model += rest[i,d,'祝休'] == 0, f"NoHolidayRest_{i}_{d}"
```

```python
# 祝日に非番の指定
for d in D_holiday:
    for i in I:
        #非番の指定があれば入れられる。
        if (i,d) in pre_dict_rest and pre_dict_rest[(i,d)] == '非番':
            continue
        #指定がなければ祝日は "出勤 or 祝休" のどちらか
        else:
            model += y[(i,d)] + rest[(i,d,'祝休')] == 1
```

```python
#"""連勤に関する制約"""
#10連勤の禁止　廃休、マル超を取れば９連勤まで可
for i in I:
    # d_idxを 0..(28-10) までループ => 0..18
    for d_idx in range(len(D) - 10 + 1):  # 28-10+1=19
        # d_idx..d_idx+9 (10日分)
        consecutive_days = [D[d_idx + offset] for offset in range(10)]
        
        # y[(i, day)] を連続10日分集計
        model += (
            pulp.lpSum(y[(i, day_)] for day_ in consecutive_days)
            <= 9
        ), f"No10ConsecutiveDays_{i}_{d_idx}"
                
#紐付け制約: normal_work <= y, normal_work <= 1 - x_special
for i in I:
    for d in D:
        model += normal_work[(i,d)] >= y[(i,d)] - specialWork[(i,d,'廃休')] - specialWork[(i,d,'マル超')]
        model += normal_work[(i,d)] <= 1 - specialWork[(i,d,'廃休')]
        model += normal_work[(i,d)] <= 1 - specialWork[(i,d,'マル超')]
        

#廃休、マル超を取らない連勤は５まで。
for i in I:
    for start_idx in range(len(D) - 6 + 1):  # 0..22
        consecutive_dates = [D[start_idx + r] for r in range(6)]
        # ここで normal_work[(i, consecutive_dates[r])] を合計
        model += (
            pulp.lpSum(normal_work[(i, d_)] for d_ in consecutive_dates)
            <= 5
        )
```

```python
LEAVE_DEDUCT = {
    '計年','年休','夏期','冬期','代休','承欠','休職','産休','育休','介護','病休','その他'
}
SPECIAL_ADD = {'廃休','マル超'}

BASE = 20
ADJUST_HELPER = 20

# 0) 日ごとの “働ける” を作る（ここを一本化）
def is_workable(i, d):
    # 週休/非番/祝休など「休み系」は False
    if (i, d) in pre_dict_rest:
        r = pre_dict_rest[(i, d)]
        if r in LEAVE_DEDUCT or r in {'非番','週休','祝休'}:
            return False
    return True

# 1) ベース供給（上限20）
employee_possible = {
    i: min(sum(1 for d in D if is_workable(i, d)), BASE)
    for i in I
}

# 2) 廃休/マル超（追加出勤扱い）
for i in I:
    add = sum(
        1 for d in D
        if (i, d) in pre_dict_special and pre_dict_special[(i, d)] in SPECIAL_ADD
    )
    employee_possible[i] += add

# 3) reqにない業務の事前指定（その日は埋まるので -1）
deduct_days = set()
for (i, d), k in pre_dict_work.items():
    if (d, k) not in req:
        deduct_days.add((i, d))
for (i, d) in deduct_days:
    employee_possible[i] -= 1

# 4) 需要
raw_demand = sum(req.values())
total_demand = raw_demand - ADJUST_HELPER

# 5) 合計
total_supply = sum(employee_possible.values())
excess = total_supply - total_demand

print("=== 社員ごとの出勤可能日数（供給） ===")
for i in sorted(employee_possible.keys()):
    print(f"{i:>6} : {employee_possible[i]:>3} 日")

print("-" * 40)
print(f"total_supply = {total_supply}")

print("\n=== 需要（req）の内訳 ===")
print(f"raw_demand (sum(req.values())) = {raw_demand}")
print(f"adjust_helper               = {ADJUST_HELPER}")
print(f"total_demand                = {total_demand}")

print("\n=== 供給 − 需要（過不足） ===")
sign = "+" if excess > 0 else ""
print(f"excess = {sign}{excess}")

if excess < 0:
    print("⚠️ 供給不足です（シフト成立不可の可能性あり）")
elif excess == 0:
    print("◎ 供給と需要が一致しています")
else:
    print("◯ 供給に余裕があります")

print("\n=== 調整要因サマリ ===")

# 廃休・マル超の合計
total_special_add = sum(
    1 for (i, d), v in pre_dict_special.items()
    if v in SPECIAL_ADD
)
print("廃休/マル超 合計日数 =", total_special_add)

# reqにない業務で引いた日数
print("reqにない業務で減算した日数 =", len(deduct_days))

# 休暇指定合計
print("休暇指定日数 =", len(pre_dict_rest))

```

```python
# =========================
#  需要(demand) 日別：その日に稼働する区数
# =========================
# zones.csv の曜日列（想定）
DAYCOLS = ["月","火","水","木","金","土","日","祝"]

missing = [c for c in DAYCOLS if c not in df_zone.columns]
if missing:
    print("zones.csv columns =", df_zone.columns.tolist())
    raise ValueError(f"zones.csv に曜日列がありません: missing={missing}")

# 日付→曜日ラベル（祝なら"祝"、それ以外は月火水...、日曜は"日"）
JPWEEK = ["月","火","水","木","金","土","日"]

def day_label(d: pd.Timestamp) -> str:
    if jpholiday.is_holiday(d):
        return "祝"
    return JPWEEK[d.weekday()]  # 0=月…6=日

# zones.csv から「その日に稼働する区」を返す
def active_zones_for_day(lbl: str) -> list[str]:
    # 稼働が数値(10/0)でも文字でも、とにかく「0/空/NaN以外」を稼働扱いにする
    s = df_zone[[zone_name_col, lbl]].copy()
    s[zone_name_col] = s[zone_name_col].astype(str).str.strip()
    col = s[lbl]

    def is_active(v):
        if pd.isna(v):
            return False
        vstr = str(v).strip()
        if vstr == "" or vstr == "0":
            return False
        # "0.0"も除外
        try:
            if float(vstr) == 0.0:
                return False
        except:
            pass
        return True

    act = s.loc[col.map(is_active), zone_name_col].tolist()
    return act

# 日別 demand（区数）
demand_by_date = {}
active_zones_by_date = {}

for d in D:
    lbl = day_label(d)
    zones = active_zones_for_day(lbl)
    active_zones_by_date[d] = zones
    demand_by_date[d] = len(zones)

print("=== 日別 需要(demand) ===")
for d in D:
    print(d.strftime("%Y-%m-%d"), day_label(d), "demand=", demand_by_date[d])

print("total_demand =", sum(demand_by_date.values()))
```

```python
# =========================
#  供給(supply) 日別：出勤できる人数
# =========================
LEAVE_BLOCK = {
    "非番","週休","祝休",
    "計年","年休","夏期","冬期","代休","承欠","休職","産休","育休","介護","病休","その他"
}
SPECIAL_ADD = {"廃休","マル超"}  # 供給+1扱い（追加出勤）

# 全社員集合 I（emp_no）
I = sorted(df_emp[emp_no_col].astype(str).str.strip().unique().tolist()) if emp_no_col else sorted({k[0] for k in pre_dict_time.keys()} | {k[0] for k in pre_dict_rest.keys()} | {k[0] for k in pre_dict_work.keys()} | {k[0] for k in pre_dict_special.keys()})

def is_off(emp: str, d: pd.Timestamp) -> bool:
    # 下段に休みコードが入っていれば休み扱い
    v = pre_dict_rest.get((emp, d))
    if v is None:
        return False
    return v in LEAVE_BLOCK

# 日別供給：休みでない人を数える
supply_by_date = {}
off_count_by_date = {}
special_add_by_date = {}

for d in D:
    off = 0
    add = 0
    for emp in I:
        if is_off(emp, d):
            off += 1
        # 廃休/マル超は“追加出勤枠”として数える（あなたのロジック踏襲）
        if pre_dict_special.get((emp, d)) in SPECIAL_ADD:
            add += 1

    # 基本供給：社員数 - 休み人数
    base_supply = len(I) - off

    # 追加供給：廃休/マル超
    supply_by_date[d] = base_supply + add
    off_count_by_date[d] = off
    special_add_by_date[d] = add

print("=== 日別 供給(supply) ===")
for d in D:
    print(
        d.strftime("%Y-%m-%d"), day_label(d),
        "supply=", supply_by_date[d],
        "(off=", off_count_by_date[d], "add=", special_add_by_date[d], ")"
    )

print("total_supply =", sum(supply_by_date.values()))
```
